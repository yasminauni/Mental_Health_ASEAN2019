```python
# chatbot benchmarking from https://github.com/Jabir0/Machine-Learning-GIZI/tree/main/Side%20Quest%20model%20(finetuning%20gemini%201.0%20in%20Vertex%20AI%20Studio) to make chatbot, add data from https://medium.com/@airimatsui09/which-happy-country-or-not-in-asean-ce435d0afb44
import torch
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM
import time

# Install necessary libraries (if not already installed)
# !pip install transformers torch sentencepiece

# Load the pre-trained model (replace with your actual model path)
model_name = "google/flan-t5-base"  # Example, replace with your fine-tuned model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Function to generate a response based on the input text
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(**inputs)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# ASEAN country data (replace with data from the Medium article)
country_data = {
    "Singapore": {"Happiness": 8, "Economic_Challenge": 2, "Ranking of Mental Health": 1},
    "Thailand": {"Happiness": 7, "Economic_Challenge": 4, "Ranking of Mental Health": 2},
    "Philippines": {"Happiness": 6, "Economic_Challenge": 5, "Ranking of Mental Health": 3},
    "Malaysia": {"Happiness": 7, "Economic_Challenge": 3, "Ranking of Mental Health": 4},
    "Indonesia": {"Happiness": 6, "Economic_Challenge": 6, "Ranking of Mental Health": 5},
    "Vietnam": {"Happiness": 6, "Economic_Challenge": 4, "Ranking of Mental Health": 6},
    "Myanmar": {"Happiness": 5, "Economic_Challenge": 7, "Ranking of Mental Health": 7},
    "Cambodia": {"Happiness": 5, "Economic_Challenge": 8, "Ranking of Mental Health": 8},
    "Laos": {"Happiness": 4, "Economic_Challenge": 9, "Ranking of Mental Health": 9},
    "Brunei": {"Happiness": 7, "Economic_Challenge": 1, "Ranking of Mental Health": 10},
}


# Get user input for country
country_name = (input("Enter an ASEAN country: "))

# You can then use the variable as usual
print(f"You entered: {country_name}")
print(f"The type is: {type(country_name)}")

# Check if the country is in the dataset
if country_name not in country_data:
    print(f"Data for {country_name} not found in the dataset.")
else:
  # Construct the prompt for the chatbot
  prompt = f"""Analyze the mental health's rank outlook in {country_name} given its happiness score, economic challenges, and ranking compared to other ASEAN countries.

  Consider the following data for ASEAN countries to rank mental health condition:
  {country_data}

  Happiness scores are on a scale of 1-10. Economic challenges are on a scale of 1-10 (1 = minimal, 10 = severe).  Ranking indicates position relative to other nations in terms of happiness.  Provide a mental health ranking (1-10, 1 being worst mental health, 10 being best), a percentage estimate for the likelihood of facing mental health challenges, and a justification for your response.  Consider factors such as economic hardship, social support, and access to healthcare specific to {country_name}.
  """

  # Benchmarking
  start_time = time.time()
  response = generate_response(prompt)
  end_time = time.time()
  elapsed_time = end_time - start_time

  print(response)
  print(f"\nBenchmarking:\nResponse time: {elapsed_time:.4f} seconds")
```
